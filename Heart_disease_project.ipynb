# **Splitting the Features And Target**

```python
df_encoded = pd.get_dummies(df, drop_first=True)
X = df_encoded.drop(columns='target', axis=1)
Y = df_encoded['target']
```

```python
print(X)
```

```python
print(Y)
```

# **Splitting the Dataset Into The Train Set And The Test Set**

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

# **Model Training**

```python
model = LogisticRegression(solver='liblinear')
```

# **Logistic Regression**

```python
model.fit(X_train, Y_train)
```

# **Training the Logistic Regression Model Into Train Set**

# **Accuracy Score of Training Data**

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print(f'Accuracy on Training data: {training_data_accuracy:.2f}')
```

# **Training the Logistic Regression Model Into Test Set**

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

# **Accuracy Score of Test Data**

```python
print('Accuracy on Test data :', test_data_accuracy)
```

# **Building A Predictive System**

```python
original_columns = ['age', 'sex', 'chest_pain_type', 'resting_blood_pressure', 'cholestoral', 'fasting_blood_sugar', 'rest_ecg', 'Max_heart_rate', 'exercise_induced_angina', 'oldpeak', 'slope', 'vessels_colored_by_flourosopy', 'thalassemia']

input_data = (4,0,1,130,204,0,0,172,0,1.4,2,0,2)

# Create a DataFrame from the input data
input_df = pd.DataFrame([input_data], columns=original_columns)

# Define mappings for categorical features
sex_map = {0: 'Female', 1: 'Male'}
chest_pain_type_map = {0: 'Asymptomatic', 1: 'Atypical Angina', 2: 'Non-Anginal Pain', 3: 'Typical Angina'}
fasting_blood_sugar_map = {0: 'False', 1: 'True'}
rest_ecg_map = {0: 'Normal', 1: 'ST-T Wave Abnormality', 2: 'LV Hypertrophy'}
exercise_induced_angina_map = {0: 'No', 1: 'Yes'}
slope_map = {0: 'Upsloping', 1: 'Flat', 2: 'Downsloping'}
vessels_colored_by_flourosopy_map = {0: '0', 1: '1', 2: '2', 3: '3', 4: '4'}
thalassemia_map = {0: 'Normal', 1: 'Fixed Defect', 2: 'Reversable Defect'}

# Apply mappings to the input DataFrame
input_df['sex'] = input_df['sex'].map(sex_map)
input_df['chest_pain_type'] = input_df['chest_pain_type'].map(chest_pain_type_map)
input_df['fasting_blood_sugar'] = input_df['fasting_blood_sugar'].map(fasting_blood_sugar_map)
input_df['rest_ecg'] = input_df['rest_ecg'].map(rest_ecg_map)
input_df['exercise_induced_angina'] = input_df['exercise_induced_angina'].map(exercise_induced_angina_map)
input_df['slope'] = input_df['slope'].map(slope_map)
input_df['vessels_colored_by_flourosopy'] = input_df['vessels_colored_by_flourosopy'].map(vessels_colored_by_flourosopy_map)
input_df['thalassemia'] = input_df['thalassemia'].map(thalassemia_map)

# Apply one-hot encoding
input_data_encoded = pd.get_dummies(input_df, drop_first=True)

# Align columns with the training data (X.columns)
aligned_input_data = input_data_encoded.reindex(columns=X.columns, fill_value=0)

prediction = model.predict(aligned_input_data)
print(prediction)

if (prediction[0]==0):
    print('The Person does not have a Heart Disease')
else:
    print('The Person have a Heart Disease')
```
