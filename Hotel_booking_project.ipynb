# **Data Visualization through Barplot**

```python
plt.figure(figsize=(5,4))
plt.title("Reservation Status Count")
plt.bar(["Not Cancelled","Cancelled"],df['is_canceled'].value_counts(), edgecolor ='k', width =0.7)
plt.show()
```

# **Data Visualization through Barplot on reservation in different hotels**

```python
plt.figure(figsize=(8,4))
ax1= sns.countplot(x = 'hotel', hue ='is_canceled', data = df, palette='Blues')
legend_labels= ax1.get_legend_handles_labels()
plt.title("Reservation Status in different hotels", size=20)
plt.xlabel("hotel")
plt.ylabel("number of reservations")
plt.show()
```

```python
resort_hotel = df[df['hotel'] =='Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize = True)
```

```python
city_hotel = df[df['hotel'] =='City Hotel']
city_hotel['is_canceled'].value_counts(normalize = True)
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')['adr'].mean()
city_hotel = city_hotel.groupby('reservation_status_date')['adr'].mean()
```

# **Data Visualization through Line Chart**

```python
plt.figure(figsize=(20,8))
plt.title("Avarage Daily Rate in City and Resort Hotel", fontsize = 30)
plt.plot(resort_hotel.index, resort_hotel.values, label = 'Resort Hotel')
plt.plot(city_hotel.index, city_hotel.values, label = 'City Hotel')
plt.legend(fontsize = 20)
plt.show()
```

# **Barplot is used to check reservation status in different months**

```python
df['month'] = df['reservation_status_date'].dt.month
plt.figure(figsize=(12,6))
ax1= sns.countplot(x = 'month', hue ='is_canceled', data = df, palette='Blues')
legend_labels= ax1.get_legend_handles_labels()
plt.title("Reservation Status in different months", size=20)
plt.xlabel("Month")
plt.ylabel("number of reservations")
plt.show()
```

```python
plt.figure(figsize=(15,8))
plt.title("ADR per months", fontsize = 20)
sns.barplot(x = 'month', y = 'adr', data = df)
plt.legend(fontsize = 20)
plt.show()
```

# **Data Visualization through Pie Chart**

```python
cancelled_data = df[df['is_canceled'] == 1]
top_10_country = cancelled_data['country'].value_counts()[:10]
plt.figure(figsize= (8,8))
plt.title(" Top 10 countries with reservation cancelled")
plt.pie(top_10_country, autopct = '%1.2f%%', labels = top_10_country.index)
plt.show()
```

# **Data Visualization through Line Chart**

```python
canceled_df_adr = cancelled_data.groupby('reservation_status_date')[['adr']].mean()
canceled_df_adr.reset_index(inplace= True)
canceled_df_adr.sort_values('reservation_status_date', inplace = True)

not_cancelled_data = df[df['is_canceled'] == 0]
not_cancelled_data_adr = not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_data_adr.reset_index(inplace = True)
not_cancelled_data_adr.sort_values('reservation_status_date', inplace = True)

plt.figure(figsize=(20,6))
plt.title("Avarage Daily Rate")
plt.plot(not_cancelled_data_adr['reservation_status_date'], not_cancelled_data_adr['adr'], label = 'not_cancelled')
plt.plot(canceled_df_adr['reservation_status_date'],canceled_df_adr['adr'], label = 'cancelled')
plt.legend()
```
